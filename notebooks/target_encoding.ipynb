# Target Encoding

```python
# Imports

import pandas as pd
from category_encoders import TargetEncoder
```

```python
train_set = pd.read_csv('../data/training_set.csv')
test_set = pd.read_csv('../data/test_set.csv')
```

```python
# Initialize TargetEncoder
target_encoder = TargetEncoder(cols=['body_type','city','engine_type','frame_damaged','fuel_type','has_accidents','isCab','is_new','make_name','maximum_seating','model_name','theft_title','transmission_display','wheel_system','trim_name'])

# Fit and transform training data
train_encoded = target_encoder.fit_transform(train_set, train_set['price'])

# Transform test data
test_encoded = target_encoder.transform(test_set)

print(train_encoded['model_name'][0], train_encoded['trim_name'][0])
print(test_encoded['model_name'][0], test_encoded['trim_name'][0])
```
