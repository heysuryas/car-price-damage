# Correlation Matrix

```python
# Imports

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_encoded = pd.read_csv('../data/training_set_target_encoded.csv')
test_encoded = pd.read_csv('../data/test_set_target_encoded.csv')
```

```python
correlation_matrix = train_encoded.corr()
 
# Filter the correlation values
filtered_correlation1 = correlation_matrix['price'][(correlation_matrix['price'] >= 0.05) | (correlation_matrix['price'] <= -0.05)]
```

```python
correlation_matrix = test_encoded.corr()
 
# Filter the correlation values
filtered_correlation2 = correlation_matrix['price'][(correlation_matrix['price'] >= 0.05) | (correlation_matrix['price'] <= -0.05)]
 
```

```python
result_side_by_side = pd.concat([filtered_correlation1, filtered_correlation2], axis=1)

# Display the result
print(result_side_by_side)
```

```python
# Visualise the correlation using a heatmap
plt.figure(figsize=(48, 40))

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5, annot_kws={"size": 20})
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Show the plot
plt.show()
```
