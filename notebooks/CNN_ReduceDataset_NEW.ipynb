### The below code combines the images from test, train, validation folders into a single folder for each class of image.

```python
import os
import shutil
import random

def move_and_rename_images(base_dirs,target_dir):
    
    counter = 1

    for base_dir in base_dirs:
        for sub_dir in os.listdir(base_dir):
            path = os.path.join(base_dir, sub_dir)
            if os.path.isdir(path):
                for category_dir in os.listdir(path):
                    category_path = os.path.join(path, category_dir)
                    if os.path.isdir(category_path):
                        target_category_path = os.path.join(target_dir, category_dir)
                        if not os.path.exists(target_category_path):
                            os.makedirs(target_category_path)

                        for file in os.listdir(category_path):
                            file_path = os.path.join(category_path, file)
                            if os.path.isfile(file_path):
                                extension = os.path.splitext(file)[1]
                                new_file_name = f"image_{counter}{extension}"
                                shutil.move(file_path, os.path.join(target_category_path, new_file_name))
                                counter += 1
                        shutil.rmtree(category_path)  # Remove the source folder
                    counter=0
                if not os.listdir(path):  # If the parent directory is empty, remove it
                    shutil.rmtree(path)

def equalize_image_count(target_dir):
    # Get the number of images in each subdirectory
    image_counts = {}
    for sub_dir in os.listdir(target_dir):
        sub_dir_path = os.path.join(target_dir, sub_dir)
        if os.path.isdir(sub_dir_path):
            image_counts[sub_dir] = len([f for f in os.listdir(sub_dir_path) if os.path.isfile(os.path.join(sub_dir_path, f))])

    print(image_counts)
    # Find the minimum number of images
    min_count = min(image_counts.values(), default=0)

    # Randomly delete files from other folders to match the min_count
    for sub_dir in image_counts.keys():
        sub_dir_path = os.path.join(target_dir, sub_dir)
        files = [f for f in os.listdir(sub_dir_path) if os.path.isfile(os.path.join(sub_dir_path, f))]
        if len(files) > min_count:
            files_to_delete = random.sample(files, len(files) - min_count)
            for file in files_to_delete:
               os.remove(os.path.join(sub_dir_path, file))
               print("deleting",os.path.join(sub_dir_path, file))


base_dirs_damage = ['/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/damage']
target_dir_damage = '/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/damage'

# Move and rename images
move_and_rename_images(base_dirs_damage,target_dir_damage)

# Equalize the number of images in each folder
equalize_image_count(target_dir_damage)

base_dirs_body = ['/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/body']
target_dir_body = '/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/body'

# Move and rename images
move_and_rename_images(base_dirs_body,target_dir_body)

# Equalize the number of images in each folder
equalize_image_count(target_dir_body)

```

## Delete Random Images from body folder to reduce the number of images

```python
# Only deleting images from body folder as it has the most number of images
import os
import random
def deleteRandomImages(parentDirDelete, deletePerc):

    for subdir, dirs, files in os.walk(parentDirDelete):
        totalFiles = len(files)
        numToDelete = int(totalFiles * (deletePerc / 100.0))

        # Randomly select files to delete
        filesToDelete = random.sample(files, numToDelete)

        for file in filesToDelete:
            file_path = os.path.join(subdir, file)
            if os.path.isfile(file_path):
                os.remove(file_path)

# Set the root folder path
parentDirDelete = '/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/body'  # Change this to your actual folder path

# Call the function to delete 50% of images from each subfolder
deleteRandomImages(parentDirDelete, 50)
```

## Resizing the images in both folders to 128 by 128 to reduce sizes

```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt

def loadAndResizeImage(path, new_size=(128, 128)):
    """
    path: Path to the image file.
    new_size: New size to which the image is to be resized.
    return: Resized image.
    """
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.resize(img, new_size)
    return img

def resizeImagesInFolder(parentDirDelete, new_size=(128, 128)):
    """
    root_folder: The path to the root folder containing the images.
    new_size: The new size of the images, default is 256x256.
    """
    for subdir, dirs, files in os.walk(parentDirDelete):
        for file in files:
            filepath = os.path.join(subdir, file)
            try:
                resized_img = loadAndResizeImage(filepath, new_size)
                # Saving the image using matplotlib.pyplot, as TensorFlow doesn't have a direct image saving method
                plt.imsave(filepath, resized_img.numpy().astype("uint8"))
            except Exception as e:
                print(f"Cannot resize {filepath}: {e}")

# Set the root folder path
damageDirDelete = '/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/damage'  # Change this to your actual folder path
bodyDirDelete = '/Users/sachin/My Files/Data Science/Playground/latest/group-coursework-sa3n/data/CNN/body'
# Call the function to resize images in each subfolder
resizeImagesInFolder(damageDirDelete, (128, 128))
resizeImagesInFolder(bodyDirDelete, (128, 128))

# Since we cannot perform file operations in this environment, this is a sample code.
# You should run this code in your local environment where the actual folders are present.
```
